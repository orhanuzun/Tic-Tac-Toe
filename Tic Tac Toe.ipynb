```python
from random import randint

def display_board(board):
    print("\n"*10)
    print(" " + board[7] + "|"+ " " + board[8] + "|" + " " + board[9])
    print("---------")
    print(" " + board[4] + "|"+ " " + board[5] + "|" + " " + board[6])
    print("---------")
    print(" " + board[1] + "|"+ " " + board[2] + "|" + " " + board[3])


def player_input():
    Sign = ""
    while Sign != "X" and Sign != "O":
        Sign = input("Player1! Please select X or O:").upper()
        if Sign == "X":
            return("X","O")
        elif Sign == "O":
            return("O","X")
def placing_sign(board,Sign, position):
    
    board[position] = Sign



def win_con(board, Sign):
    for i in range (0,3):
        if (board[i + 1] == board[i + 2] == board[i + 3] == Sign):
            return True
    for j in range(0,3):
        if (board[j + 1] == board[j + 4] == board[j + 7] == Sign):
            return True
        
    if ((board[1] == board[5] == board[9]== Sign) or (board[3] == board[5] == board[7] == Sign)):
        return True
        
        
def coinflip():  #to decide who starts
    flip = randint(0,1)
    if flip == 1:
        return "Player 1"
    elif flip == 0:
        return "player 2"
def spacecheck(board,position): #checks if the table is full
    return board[position] == " "
def full_board_check(board):
    for i in range(1,10):
        if spacecheck(board,i):
            return False
    return True 
def player_choice(board):   #which location that player gonna place their sign on 
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not spacecheck(board,position):
        position = int(input("Choose a position 1-9"))
    return position

while True:
    the_board = [" "]*10
    player1_sign, player2_sign = player_input()
    turn = coinflip()
    print("first player that is playing is :" + turn)
    game_start = input("are you ready to play Y/N ?").upper()
    if game_start == "Y":
        game_on = True
    elif game_start == "N":
        game_on = False
    while game_on:
        if turn == "Player 1":
            display_board(the_board)
            position = player_choice(the_board)
            placing_sign(the_board, player1_sign, position)
            print(the_board)
            if win_con(the_board, player1_sign):
                display_board(the_board)
                print("PLAYER 1 HAS WON THE GAME !!! ")
                game_on = False
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("THE GAME IS TIE -__-")
                    break
                else:
                    turn = "Player 2"
        else:
            display_board(the_board)
            position = player_choice(the_board)
            placing_sign(the_board, player2_sign, position)
            if win_con(the_board, player2_sign):
                display_board(the_board)
                print("PLAYER 2 HAS WON THE GAME !!! ")
                game_on = False
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print("THE GAME IS TIE")
                    break
                else:
                    turn = "Player 1"
```
